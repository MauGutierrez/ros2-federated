```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "../results/history/"
obj1 = pd.read_pickle(path + "client_1.pickle")
obj2 = pd.read_pickle(path + "client_2.pickle")
```

```python
accuracy1 = []
val_accuracy1 = []
loss1 = []
val_loss1 = []

for element in obj1:
    accuracy1 += element['accuracy']
    val_accuracy1 += element['val_accuracy']
    loss1 += element['loss']
    val_loss1 += element['val_loss']
```

```python
accuracy2 = []
val_accuracy2 = []
loss2 = []
val_loss2 = []

for element in obj2:
    accuracy2 += element['accuracy']
    val_accuracy2 += element['val_accuracy']
    loss2 += element['loss']
    val_loss2 += element['val_loss']
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,10))
fig.tight_layout(pad=5.0)
axs[0,0].plot(accuracy1)
axs[0,0].plot(val_accuracy1)
axs[0,0].set_title('Agent 1 accuracy')
axs[0,0].set(xlabel='epoch', ylabel='accuracy')
axs[0,0].legend(['train', 'val'], loc='upper left')

axs[0,1].plot(loss1)
axs[0,1].plot(val_loss1)
axs[0,1].set_title('Agent 1 loss')
axs[0,1].set(xlabel='epoch', ylabel='loss')
axs[0,1].legend(['train', 'val'], loc='upper left')

axs[1,0].plot(accuracy2)
axs[1,0].plot(val_accuracy2)
axs[1,0].set_title('Agent 2 accuracy')
axs[1,0].set(xlabel='epoch', ylabel='accuracy')
axs[1,0].legend(['train', 'val'], loc='upper left')

axs[1,1].plot(loss2)
axs[1,1].plot(val_loss2)
axs[1,1].set_title('Agent 2 loss')
axs[1,1].set(xlabel='epoch', ylabel='loss')
axs[1,1].legend(['train', 'val'], loc='upper left')

fig.savefig('agents_accuracy_loss.png')
```

```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import re

path = "../results/"
subdirectories = sorted(os.listdir(path))
reTimestamp = r'\b(\d+\.\d{,})\b'

times = {}
for name in subdirectories:
    subpath = path + name + "/time/"
    files = os.listdir(subpath)
    times[name] = {}
    for file in files:
        my_log = subpath + file
        with open(my_log) as f:
            for line in f:
                result = re.findall(reTimestamp, line)
                if result:
                    agent = file.split(".")[0]
                    times[name][agent] = float(result[0])

df = pd.DataFrame(times)

mean_times = df.mean(axis=0)
mean_times = pd.DataFrame(mean_times)
mean_times = mean_times.reset_index()
mean_times_1 = mean_times.rename(columns={'index': 'Experiment Name', 0: 'Average Communication Time (ms)'})

one_agent = mean_times_1[mean_times_1["Experiment Name"].str.contains("1_agent")]
one_agent = one_agent.sort_values(by=['Average Communication Time (ms)'])

two_agent = mean_times_1[mean_times_1["Experiment Name"].str.contains("2_agent")]
two_agent = two_agent.sort_values(by=['Average Communication Time (ms)'])

three_agent = mean_times_1[mean_times_1["Experiment Name"].str.contains("3_agent")]
three_agent = three_agent.sort_values(by=['Average Communication Time (ms)'])

four_agent = mean_times_1[mean_times_1["Experiment Name"].str.contains("4_agent")]
four_agent = four_agent.sort_values(by=['Average Communication Time (ms)'])

# one_agent = mean_times_1[mean_times_1["Experiment Name"].str.contains("1_agent")]
# one_agent = one_agent.sort_values(by=['Average Communication Time (ms)'])

# one_agent = mean_times_1[mean_times_1["Experiment Name"].str.contains("1_agent")]
# one_agent = one_agent.sort_values(by=['Average Communication Time (ms)'])

x_1 = one_agent['Average Communication Time (ms)'].values
y_1 = one_agent['Experiment Name'].values
y_1 = [int(experiment.split("_")[2]) for experiment in y_1]

x_2 = two_agent['Average Communication Time (ms)'].values
y_2 = two_agent['Experiment Name'].values
y_2 = [int(experiment.split("_")[2]) for experiment in y_2]

x_3 = three_agent['Average Communication Time (ms)'].values
y_3 = three_agent['Experiment Name'].values
y_3 = [int(experiment.split("_")[2]) for experiment in y_3]

x_4 = four_agent['Average Communication Time (ms)'].values
y_4 = four_agent['Experiment Name'].values
y_4 = [int(experiment.split("_")[2]) for experiment in y_4]

print(x_1, x_2, x_3, x_4)

plt.plot(x_1, y_1, color='red', marker='o')
plt.plot(x_2, y_2, color='green', marker='o')
plt.plot(x_3, y_3, color='blue', marker='o')
plt.plot(x_4, y_4, color='orange', marker='o')
plt.yticks([1, 5, 10, 15, 20])
plt.title('Federated Server/Agents Computation time')
plt.xlabel('Computation time (sec)')
plt.ylabel('Iterations')
plt.legend(['1 agent', '2 agents', '3 agents', '4 agents'], loc='lower right')
plt.savefig('computation_time.png')
plt.show()
```

```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import re

paths = ["../results/1_agent_1_iterations_20_epoch/history/", "../results/2_agent_1_iterations_20_epoch/history/",
          "../results/3_agent_1_iterations_20_epoch/history/", "../results/4_agent_1_iterations_20_epoch/history/"]

experiments = {}
for path in paths:
    subdirs = os.listdir(path)
    acc = np.zeros(21)
    for file in subdirs:
        temp = []
        obj1 = pd.read_pickle(path+file)
        for element in obj1:
            temp += element['val_accuracy']
        temp = np.array(temp)
        acc += temp
        
    agents = len(subdirs)
    acc /= agents
    experiments[agents] = acc
    
plt.plot(experiments[1])
plt.plot(experiments[2])
plt.plot(experiments[3])
plt.plot(experiments[4])
plt.xticks(range(0, 21, 2))
plt.title('Average Test accuracy')
plt.xlabel('1 iteration 20 epochs')
plt.ylabel('accuracy')
plt.legend(['1 agent', '2 agents', '3 agents', '4 agents'], loc='lower right')
plt.savefig('1_iteration_20_epochs_acc.png')
plt.close()
```
